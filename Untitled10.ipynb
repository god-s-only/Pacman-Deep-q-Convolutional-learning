```python
class Network(nn.Module):

  def __init__(self, action_size, seed = 42):
    super(Network, self).__init__()
    self.seed = torch.manual_seed(seed)
    self.conv1 = nn.Conv2d(3, 32, kernel_size = 8, stride = 4)
    self.bn1 = nn.BatchNorm2d(32)
    self.conv2 = nn.Conv2d(32, 64, kernel_size = 4, stride = 2)
    self.bn2 = nn.BatchNorm2d(64)
    self.conv3 = nn.Conv2d(64, 64, kernel_size = 3, stride = 1)
    self.bn3 = nn.BatchNorm2d(64)
    self.conv4 = nn.Conv2d(64, 128, kernel_size = 3, stride = 1)
    self.bn4 = nn.BatchNorm2d(128)
    self.fc1 = nn.Linear(10 * 10 * 128, 512)
    self.fc2 = nn.Linear(512, 256)
    self.fc3 = nn.Linear(256, action_size)

  def forward(self, state):
    x = F.relu(self.bn1(self.conv1(state)))
    x = F.relu(self.bn2(self.conv2(x)))
    x = F.relu(self.bn3(self.conv3(x)))
    x = F.relu(self.bn4(self.conv4(x)))
    x = x.view(x.size(0), -1)
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    return self.fc3(x)
```

```python
import gym

env = gym.make("MsPacman-v0", render_mode="rgb_array")
obs = env.reset()
state_shape = env.observation_space.shape
state_size = env.observation_space.shape[0]
number_actions = env.action_space.n
print("Observation shape:", obs[0].shape)  # obs is (obs, info)
print("Action space size:", env.action_space.n)
```

```python
learning_rate = 5e-4
minibatch_size = 64
discount_factor = 0.99
```

```python
from PIL import Image
from torchvision import transforms

def preprocess_frame(frame):
    frame = Image.fromarray(frame)
    preprocess = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor()  # Automatically converts to [C, H, W] format
    ])
    return preprocess(frame)  # returns shape [3, 128, 128]
```

```python
class Agent:
    def __init__(self, action_size):
        self.device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
        self.local_q = Network(action_size).to(self.device)
        self.target_q = Network(action_size).to(self.device)
        self.optimizer = optim.Adam(self.local_q.parameters(), lr=learning_rate)
        self.memory = deque(maxlen=10000)
        self.action_size = action_size

    def step(self, state, action, reward, next_state, done):
        state = preprocess_frame(state).cpu()
        next_state = preprocess_frame(next_state).cpu()
        self.memory.append((state, action, reward, next_state, done))
        if len(self.memory) > minibatch_size:
            batch = random.sample(self.memory, minibatch_size)
            self.learn(batch)

    def act(self, state, epsilon=0.):
        state = preprocess_frame(state).unsqueeze(0).to(self.device)
        self.local_q.eval()
        with torch.no_grad():
            q_values = self.local_q(state)
        self.local_q.train()
        if random.random() > epsilon:
            return np.argmax(q_values.cpu().numpy())
        else:
            return random.choice(np.arange(self.action_size))

    def learn(self, experiences):
        states, actions, rewards, next_states, dones = zip(*experiences)
        states = torch.stack(states).float().to(self.device)
        next_states = torch.stack(next_states).float().to(self.device)
        actions = torch.tensor(actions).unsqueeze(1).long().to(self.device)
        rewards = torch.tensor(rewards).unsqueeze(1).float().to(self.device)
        dones = torch.tensor(dones).unsqueeze(1).float().to(self.device)

        q_targets_next = self.target_q(next_states).detach().max(1)[0].unsqueeze(1)
        q_targets = rewards + (discount_factor * q_targets_next * (1 - dones))
        q_expected = self.local_q(states).gather(1, actions)

        loss = F.mse_loss(q_expected, q_targets)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
```

```python
# Step 5: Training loop
agent = Agent(number_actions)
episodes = 100
max_timesteps = 10000
epsilon = 1.0
eps_end = 0.01
eps_decay = 0.995
scores = deque(maxlen=100)

for episode in range(1, episodes + 1):
    state, _ = env.reset()
    score = 0
    for t in range(max_timesteps):
        action = agent.act(state, epsilon)
        next_state, reward, done, _, _ = env.step(action)
        agent.step(state, action, reward, next_state, done)
        state = next_state
        score += reward
        if done:
            break
    scores.append(score)
    epsilon = max(eps_end, eps_decay * epsilon)
    print(f"\rEpisode {episode}\tAverage Score: {np.mean(scores):.2f}", end="")
    if episode % 100 == 0:
        print(f"\rEpisode {episode}\tAverage Score: {np.mean(scores):.2f}")
    if np.mean(scores) >= 500.0:
        print(f"\nEnvironment solved in {episode} episodes!\tAverage Score: {np.mean(scores):.2f}")
        torch.save(agent.local_q.state_dict(), 'checkpoint.pth')
        break
```

```python
# Step 6: Record gameplay video
import glob, io, base64, imageio
from IPython.display import HTML, display

def show_video_of_model(agent, env_name="MsPacman-v0"):
    env = gym.make(env_name, render_mode='rgb_array')
    state, _ = env.reset()
    done = False
    frames = []
    while not done:
        frame = env.render()
        frames.append(frame)
        action = agent.act(state, epsilon=0.0)
        state, reward, done, _, _ = env.step(action)
    env.close()
    imageio.mimsave('video.mp4', frames, fps=30)

def show_video():
    mp4list = glob.glob('video.mp4')
    if mp4list:
        video = io.open(mp4list[0], 'r+b').read()
        encoded = base64.b64encode(video)
        display(HTML(data=f'''<video alt="video" autoplay loop controls style="height: 400px;">
            <source src="data:video/mp4;base64,{encoded.decode('ascii')}" type="video/mp4" />
        </video>'''))
    else:
        print("No video found")

# After training:
show_video_of_model(agent)
show_video()
```
